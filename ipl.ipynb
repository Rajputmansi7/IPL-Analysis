## CAGR & Future Value function


```python
# Latest revenue data from fact_revenue_demography.xlsx (FY23 or FY24)
latest_data = [
    {"Company": "Dream11", "Latest_Revenue": 6384.0, "Year": 2023, "Growth_Rate": 0.3},
    {"Company": "My11Circle", "Latest_Revenue": 2250.0, "Year": 2023, "Growth_Rate": 0.3},
    {"Company": "PokerBaazi", "Latest_Revenue": 415.0, "Year": 2024, "Growth_Rate": 0.3},
    {"Company": "Vimal Elaichi", "Latest_Revenue": 5267.0, "Year": 2024, "Growth_Rate": 0.09},
    {"Company": "Rajshree", "Latest_Revenue": 2500.0, "Year": 2024, "Growth_Rate": 0.09}
]

# Function to calculate future value
def calculate_future_value(present_value, growth_rate, years):
    return present_value * (1 + growth_rate) ** years

# Function to calculate CAGR
def calculate_cagr(start_value, end_value, periods):
    return ((end_value / start_value) ** (1 / periods) - 1) * 100
    # Formula: ((Ending Value / Starting Value)^(1/Periods) - 1) * 100
```

## CAGR 2025-2030 evaluation

```python
# Parameters
start_year = 2025
end_year = 2030
periods = end_year - start_year  # 5 years

# Calculate 2025 and 2030 revenues
results = []
companies = []
cagr_values = []
for item in latest_data:
    # Calculate years from latest year to 2025
    years_to_2025 = start_year - item["Year"]
    # Calculate 2025 revenue (Starting Value)
    revenue_2025 = calculate_future_value(item["Latest_Revenue"], item["Growth_Rate"], years_to_2025)
    # Calculate 2030 revenue (Ending Value)
    revenue_2030 = calculate_future_value(revenue_2025, item["Growth_Rate"], periods)
    # Calculate CAGR
    cagr = calculate_cagr(revenue_2025, revenue_2030, periods)
    # Store results
    results.append({
        "Company": item["Company"],
        "2025 Revenue ( Crore)": round(revenue_2025, 2),
        "2030 Revenue ( Crore)": round(revenue_2030, 2),
        "CAGR (%)": round(cagr, 2)
    })
    companies.append(item["Company"])
    cagr_values.append(round(cagr, 2))
```

## Calculating Total Users and Negatively Impacted People

```python
# Step 1: Define mapping for user base estimates
brand_user_base = {
    'Dream11': 220_000_000,
    'PokerBaazi': 5_000_000,
    'My11Circle': 40_000_000,
    'Vimal Pan Masala': None,
    'Kamla Pasand': None,
    'Rajshree': None
}

india_population = 1_400_000_000
pan_masala_prevalence = 0.12
pan_masala_users = int(india_population * pan_masala_prevalence)

def get_user_base(row):
    if row['company'] in brand_user_base and brand_user_base[row['company']]:
        return brand_user_base[row['company']]
    elif row['sector'] == 'Pan Masala':
        return pan_masala_users // 3
    else:
        return None

def get_negative_impact_pct(row):
    if row['sector'] == 'Fantasy/App':
        return 0.20
    elif row['sector'] == 'Pan Masala':
        return 0.032
    else:
        return 0

# Step 2: Calculate estimated users and negative impact percentage
df_revenue = df_revenue.copy()
df_revenue['estimated_users'] = df_revenue.apply(get_user_base, axis=1)
df_revenue['neg_impact_pct'] = df_revenue.apply(get_negative_impact_pct, axis=1)

# Step 3: Calculate negatively impacted population
df_revenue['negatively_impacted_population'] = (
    df_revenue['estimated_users'] * df_revenue['neg_impact_pct']
).astype('Int64')
```

# 4. Total Population Negatively Impacted by Pan Masala and Betting apps

```python
# Step 4: Prepare summary table - Create a new column for formatted percentage
summary = df_revenue[['company', 'sector', 'estimated_users', 'neg_impact_pct', 'negatively_impacted_population']].copy()
summary['neg_impact_pct_formatted'] = (summary['neg_impact_pct'] * 100).round(1).astype(str) + '%'

# Step 5: Aggregate by sector
sector_summary = summary.groupby('sector', as_index=False).agg({
    'estimated_users': 'sum',
    'negatively_impacted_population': 'sum'
})
sector_summary['neg_impact_pct'] = (
    sector_summary['negatively_impacted_population'] /
    sector_summary['estimated_users'] * 100
).round(1)
sector_summary['neg_impact_pct_formatted'] = sector_summary['neg_impact_pct'].astype(str) + '%'

print("Detailed Brand-level Impact:")
print(summary[['company', 'sector', 'estimated_users', 'neg_impact_pct_formatted', 'negatively_impacted_population']]
      .rename(columns={'neg_impact_pct_formatted': 'neg_impact_pct'})
      .to_string(index=False))
print("\nSector-wise Summary:")
print(sector_summary[['sector', 'estimated_users', 'negatively_impacted_population', 'neg_impact_pct_formatted']]
      .rename(columns={'neg_impact_pct_formatted': 'neg_impact_pct'})
      .to_string(index=False))
```

# Network Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a bipartite graph
G = nx.Graph()

# Add nodes with 'bipartite' attribute
celebrities = result.index.tolist()
brands_set = set()

# Build edges and collect brands
edges = []
for celeb in celebrities:
    celeb_brands = result.loc[celeb, 'brands'].split(', ')
    brands_set.update(celeb_brands)
    for brand in celeb_brands:
        edges.append((celeb, brand.strip()))

# Add nodes
G.add_nodes_from(celebrities, bipartite=0)
G.add_nodes_from(brands_set, bipartite=1)

# Add edges
G.add_edges_from(edges)

# Set positions for bipartite layout
pos = nx.bipartite_layout(G, celebrities)

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, nodelist=celebrities, node_color='lightcoral', node_size=1000, label='Celebrities')
nx.draw_networkx_nodes(G, pos, nodelist=brands_set, node_color='skyblue', node_size=900, label='Brands')
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')

plt.title("CelebrityHigh-Risk Brand Endorsement Network", fontsize=14, weight='bold')
plt.axis('off')
plt.legend(loc='center right')
plt.tight_layout()
plt.show()
```

## **Public Health Implications**

```python
# Step 1: Create a public health impact index
# We'll combine revenue, health/social risk, and demographic vulnerability to calculate an index

# Map risk levels to numeric scores
risk_scores = {
    'Extremely High - Carcinogenic (Surrogate Ads)': 10,
    'Extremely High Carcinogenic (Surrogate Ads)': 10,
    'Extremely High - Promotes Gambling': 10,
    'Extremely High Promotes Gambling': 10,
    'Carcinogenic (Surrogate Ads)': 8,
    'Moderate  (Sugar Content)': 4
}

advertisers_df=pd.read_excel('fact_ipl_advertisers.xlsx')
revenue_demo_df=pd.read_excel('fact_revenue_demography.xlsx')

# Merge advertisers and revenue demographic data on brand/company
merged_df = advertisers_df.merge(revenue_demo_df, how='left', left_on='advertiser_brand', right_on='company')

# Apply risk score
merged_df['health_risk_score'] = merged_df['health_social_risk'].map(risk_scores)
# Parse revenue like '6,200-6,500 (FY23)' into a float (e.g., 6350.0)
def parse_revenue(value):
    try:
        value = str(value).replace(',', '').replace('(FY23)', '').replace('(FY24)', '').replace('(est.)', '')
        parts = value.split('-')
        if len(parts) == 2:
            return (float(parts[0]) + float(parts[1])) / 2  # average if it's a range
        else:
            return float(parts[0])  # single value
    except:
        return None  # fallback if it can't be parsed

# Clean and parse revenue again for merged_df
merged_df['parsed_revenue'] = merged_df['latest_annual_revenue'].apply(parse_revenue)

# Define vulnerability score by urban reach and age group (simplified logic)
def calc_vulnerability(row):
    age_score = 1 if '15' in str(row['age_group']) or '18' in str(row['age_group']) else 0
    urban_score = 1 if 'Tier 2' in str(row['urban_population']) or 'Rural' in str(row['urban_population']) else 0
    return age_score + urban_score

merged_df['vulnerability_score'] = merged_df.apply(calc_vulnerability, axis=1)

# Normalize revenue (scale to 0-10 range for simplicity)
max_revenue = merged_df['parsed_revenue'].max()
merged_df['normalized_revenue'] = merged_df['parsed_revenue'] / max_revenue * 10

# Create Public Health Impact Index (PHII)
merged_df['PHII'] = merged_df['health_risk_score'].fillna(0) + \
                    merged_df['vulnerability_score'] + \
                    merged_df['normalized_revenue'].fillna(0)

# Step 2: Visualize PHII
plt.figure(figsize=(12, 6))
phii_plot = merged_df.sort_values(by='PHII', ascending=False)
sns.barplot(data=phii_plot, x='PHII', y='advertiser_brand', palette='Reds_d')
plt.title('Public Health Impact Index (PHII) of IPL Advertisers')
plt.xlabel('PHII (Higher = More Harmful)')
plt.ylabel('Advertiser Brand')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
def create_spider_chart():
    # Categories and values
    categories = ['Ethical Compliance', 'Consumer Protection',
                 'Social Responsibility', 'Transparency', 'Regulatory Adherence']
    values = [65, 60, 55, 70, 60]

    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]
    values += values[:1]

    # Dark contrasting color scheme
    fig = plt.figure(figsize=(10, 10), facecolor='#2e2e2e')  # Dark background
    ax = fig.add_subplot(111, polar=True, facecolor='#2e2e2e')

    # Plot styling
    ax.spines['polar'].set_color('white')
    ax.tick_params(axis='both', colors='white')

    # Plot data with contrasting colors
    ax.plot(angles, values, linewidth=3, linestyle='solid',
            color='#ff9e3b', marker='o', markersize=10)  # Orange line
    ax.fill(angles, values, color='#ff9e3b', alpha=0.25)

    # Axis labels - prevent overlap
    plt.xticks(angles[:-1], categories, color='white', size=12, wrap=True)
    plt.yticks([20, 40, 60, 80], ["20", "40", "60", "80"],
               color="white", size=10)
    plt.ylim(0, 100)

    # Adjust label positions
    for label, angle in zip(ax.get_xticklabels(), angles[:-1]):
        if angle == 0:
            label.set_horizontalalignment('center')
        elif 0 < angle < np.pi:
            label.set_horizontalalignment('left')
        else:
            label.set_horizontalalignment('right')

    # Title with contrast
    plt.title("IPL Advertising Ethics Index - Dimension Analysis",
              size=16, y=1.1, color='white', weight='bold')

    plt.tight_layout()
    plt.show()

create_spider_chart()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import textwrap

# Apply style gracefully
if 'seaborn-whitegrid' in plt.style.available:
    plt.style.use('seaborn-whitegrid')
else:
    plt.style.use('ggplot')

# Framework stages and strategies
stages = [
    "1. Advertiser\nSelection",
    "2. Risk\nEvaluation",
    "3. Regulation &\nMonitoring",
    "4. Content\nOptimization",
    "5. Revenue\nReinforcement",
    "6. Community\nImpact Mitigation"
]

strategies = {
    stages[0]: [
        " Prioritize FMCG, OTT brands",
        " Limit Pan Masala/Gambling ads",
        " Implement ethical screening"
    ],
    stages[1]: [
        " Use AEI risk scoring",
        " Balance reach vs social impact",
        " Quarterly brand audits"
    ],
    stages[2]: [
        " Stricter ad guidelines",
        " Mandate clear disclaimers",
        " Real-time content monitoring"
    ],
    stages[3]: [
        " More family-friendly ads",
        " Positive celebrity messaging",
        " Educational content slots"
    ],
    stages[4]: [
        " Incentivize ethical brands",
        " Revenue sharing models",
        " Premium ad slots for ESG"
    ],
    stages[5]: [
        " 2% revenue to social causes",
        " Anti-addiction campaigns",
        " Sports education programs"
    ]
}

# Figure setup
fig, ax = plt.subplots(figsize=(14, 12))
ax.axis("off")

colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948']
dark_colors = ['#2d5b8f', '#d46e0a', '#c9393b', '#569b96', '#3d8b38', '#dbb52e']
y_pos = [0.85, 0.7, 0.55, 0.4, 0.25, 0.1]
stage_box_width, stage_box_height = 0.3, 0.12

# Draw chart
for i, (stage, color, dark_color) in enumerate(zip(stages, colors, dark_colors)):
    # Main stage box
    stage_box = FancyBboxPatch((0.5 - stage_box_width / 2, y_pos[i] - stage_box_height / 2),
                               stage_box_width, stage_box_height,
                               boxstyle="round,pad=0.02,rounding_size=0.05",
                               facecolor=color, edgecolor=dark_color, lw=2)
    ax.add_patch(stage_box)
    ax.text(0.5, y_pos[i], stage, ha='center', va='center',
            fontsize=15, weight='bold', color='white')

    # Strategy box layout
    strat_x = 0.70
    strat_width = 0.28
    line_height = 0.022
    spacing = 0.01
    current_y = y_pos[i] - 0.03

    for point in strategies[stage]:
        # Wrap text manually
        wrapped = textwrap.wrap(point, width=38)  # Adjust width as needed
        box_height = line_height * len(wrapped) + 0.01

        # Draw box
        strat_box = FancyBboxPatch((strat_x, current_y - box_height / 2),
                                   strat_width, box_height,
                                   boxstyle="round,pad=0.01,rounding_size=0.02",
                                   facecolor='white', edgecolor=dark_color, lw=1.5)
        ax.add_patch(strat_box)

        # Add each line of text centered vertically
        for k, line in enumerate(wrapped):
            y_line = (current_y + box_height / 2 - 0.01) - k * line_height
            ax.text(strat_x + 0.01, y_line, line,
                    ha='left', va='top', fontsize=10.5, color='black')

        current_y -= (box_height + spacing)

    # Arrows
    if i < len(stages) - 1:
        ax.annotate("",
                    xy=(0.5, y_pos[i+1] + stage_box_height / 2 + 0.01),
                    xytext=(0.5, y_pos[i] - stage_box_height / 2 - 0.01),
                    arrowprops=dict(arrowstyle="->", color='gray', lw=2))

# Title
plt.title("IPL Advertising: Framework for Economic Growth with Social Responsibility",
          fontsize=16, weight='bold', pad=20, color='#2e2e2e')

# Metrics box
metrics = [
    "Projected Outcomes:",
    " +1520% revenue from ethical brands",
    " 30% social risk complaints",
    " 25% improvement in ESG scores",
    " 2x brand recall for FMCG sponsors"
]
ax.text(0.05, 0.02, "\n".join(metrics),
        ha='left', va='bottom', fontsize=10,
        bbox=dict(facecolor='#f7f7f7', edgecolor='#d9d9d9', boxstyle='round,pad=0.5'))

# Show plot
plt.subplots_adjust(top=0.92, bottom=0.05)
plt.show()
```
